# <div align="center"> Data Dictionary

|Columns|Discription|
|---|---|
|state|state|
|date|timestamp|
|total_doses_distributed|Total accumulated doses distributed|
|daily_change_dist|Daily change on total accumulated doses distributed|
|total_doses_administered|Total accumulated doses administered|
|daily_change_admin|Daily change on total accumulated doses administered|
|doses_distributed_100k|Doses distributed for every 100,000 ppl|
|dist_100k_change|Daily change on doses distributed for every 100,000 ppl|
|doses_administed_100k|Doses administered every 100,000 ppl|
|admin_100k_change|Daily change on doses administered every 100,000 ppl|
|people_1+_dose|People receiving 1 or more doses|
|people_1+_dose_change|Daily change on people receiving 1 or more doses|
|doses_1+_100k|How many people received 1 or More Doses for every 100,000 ppl|
|doses_1+_100k_change|Daily change on how many people received 1 or More Doses for every 100,000 ppl|
|people_2_doses|People receiving 2 or more doses|
|people_2_doses_change|Daily change on people receiving 2 or more doses|
|doses_2_100k|How many people received 2 or More Doses for every 100,000 ppl|
|doses_2_100k_change|Daily change on how many people received 2 or More Doses for every 100,000 ppl|



```python
# create a column for the daily distrubution change

vac.insert(loc=3,column='daily_change_dist',
    value=vac['total_doses_distributed']-vac['total_doses_distributed'].shift(1))
```

```python
vac.insert(loc=5,column='daily_change_admin',
    value=vac['total_doses_administered']-vac['total_doses_administered'].shift(1))
```

```python
vac.insert(loc=7,column='dist_100k_change',
    value=vac['doses_distributed_100k']-vac['doses_distributed_100k'].shift(1))
```

```python
vac.insert(loc=9,column='admin_100k_change',
    value=vac['doses_administed_100k']-vac['doses_administed_100k'].shift(1))
```

```python
vac.insert(loc=11,column='people_1+_dose_change',
    value=vac['people_1+_dose']-vac['people_1+_dose'].shift(1))
```

```python
vac.insert(loc=13,column='doses_1+_100k_change',
    value=vac['doses_1+_100k']-vac['doses_1+_100k'].shift(1))
```

```python
vac.insert(loc=15,column='people_2_doses_change',
    value=vac['people_2_doses']-vac['people_2_doses'].shift(1))
```

```python
vac.insert(loc=17,column='doses_2_100k_change',
    value=vac['doses_2_100k']-vac['doses_2_100k'].shift(1))
```

```python
vac.fillna(0,inplace=True)
```
